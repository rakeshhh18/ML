#label encoding of cancer dataset

```python
#importing the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
#loading the data from csv file to pandas dataframe
cancer=pd.read_csv('/content/data.csv')
```

```python

#printing first five rows of the dataframe
cancer.head()
```

```python
#counting the no of labels in diagnosis column
cancer['diagnosis'].value_counts()
```

```python
#load the label encoder function
label_encode=LabelEncoder()
```

```python
labels=label_encode.fit_transform(cancer.diagnosis)
```

```python
#appending the labels to the dataframe
cancer['target']=labels
```

```python
#printing first five rows
cancer.head()
```

```python
cancer.drop(columns='diagnosis',axis=1)
```
